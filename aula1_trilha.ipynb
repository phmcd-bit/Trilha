## Questão 3 📝

Dada a string abaixo, use métodos de string para:
1. Imprimir em maiúsculas
2. Imprimir em minúsculas
3. Imprimir somente os primeiros 6 caracteres (slicing)
4. Verificar se a palavra `"Python"` está contida na frase

**Código inicial:**

```bash
frase = "Python é incrível"

# 1. maiúsculas
# 2. minúsculas
# 3. primeiros 6 caracteres
# 4. verificar se "Python" está na frase
```

````python
frase = "Python é incrível"

# 1. maiúsculas
print(frase.upper())

# 2. minúsculas
print(frase.lower())

# 3. primeiros 6 caracteres
print(frase[:6])

# 4. verificar se "Python" está na frase
if "Python" in frase:
  print("'Python' está na frase.")
else:
  print("'Python' não está na frase.")
````

### 💡 Dica:

## Questão 6 📝

Crie um programa que recebe o salário de um funcionário e calcula o imposto devido com base nas faixas abaixo:

| Faixa salarial | Alíquota |
|---|---|
| Até R\$ 1.500 | Isento (0%) |
| R\$ 1.500,01 – R\$ 3.000 | 7,5% |
| R\$ 3.000,01 – R\$ 4.500 | 15% |
| Acima de R\$ 4.500 | 22,5% |

Exiba o salário bruto, o valor do imposto e o salário líquido.

````python
salario = float(input("Digite seu salário: "))

if salario <= 1500:
  aliquota = 0
  imposto = 0
elif salario <= 3000:
  aliquota = 0.075
  imposto = salario * aliquota
elif salario <= 4500:
  aliquota = 0.15
  imposto = salario * aliquota
else:
  aliquota = 0.225
  imposto = salario * aliquota

liquido = salario - imposto

print(f"Salário bruto: R$ {salario:.2f}")
print(f"Valor do imposto: R$ {imposto:.2f}")
print(f"Salário líquido: R$ {liquido:.2f}")
````

### 💡 Dica:

*Utilize somente se necessário — tente resolver primeiro sozinho!*

*Use `if`/`elif`/`else` para cada faixa. Calcule `imposto = salario * aliquota`, depois `liquido = salario - imposto`. *

## Questão 7 📝

Crie um programa que verifica se um ano digitado pelo usuário é bissexto.

**Regras:**
- Divisível por 4 → bissexto
- Mas se divisível por 100 → **NÃO** é bissexto
- A menos que também seja divisível por 400 → **É** bissexto

Exiba `"Ano bissexto"` ou `"Não é bissexto"`.

````python
ano = int(input("Digite um ano: "))

if ano % 4 == 0:
  if ano % 100 == 0:
      if ano % 400 == 0:
          print(f"O ano de {ano} é bissexto.")
      else:
          print(f"O ano de {ano} não é bissexto.")
  else:
      print(f"O ano de {ano} é bissexto.")
else:
  print(f"O ano de {ano} não é bissexto.")

````

### 💡 Dica:

*Utilize somente se necessário — tente resolver primeiro sozinho!*

*Use o operador `%` para verificar divisibilidade. Combine condições com `and` e `or`.*

## Questão 8 📝

Construa um programa que recebe o peso e a altura do usuário, calcula o IMC e classifica o resultado:

| IMC | Classificação |
|---|---|
| IMC < 18,5 | Abaixo do peso |
| 18,5 <= IMC < 25 | Peso normal |
| 25 <= IMC < 30 | Sobrepeso |
| IMC >= 30 | Obesidade |

Exiba o IMC com 2 casas decimais e a classificação.

````python
peso = float(input("Digite seu peso: "))
altura = float(input("Digite sua altura: "))

imc = peso / (altura ** 2)

if imc < 18.5:
  classificacao = "Abaixo do peso"
  print(f"Seu IMC é igual a {imc:.2f} e sua classificação é: {classificacao}.")
elif imc < 25:
  classificacao = "Peso normal"
  print(f"Seu IMC é igual a {imc:.2f} e sua classificação é: {classificacao}")
elif imc < 30:
  classificacao = "Sobrepeso"
  print(f"Seu IMC é igual a {imc:.2f} e sua classificação é: {classificacao}")
else:
  classificacao = "Obesidade"
  print(f"Seu IMC é igual a {imc:.2f} e sua classificação é: {classificacao}")
````

### 💡 Dica:

*Utilize somente se necessário — tente resolver primeiro sozinho!*

*`IMC = peso / altura**2`. Para formatar com 2 casas decimais, use `f"{imc:.2f}"`.*

## Questão 10 📝

Crie um classificador de triângulos. Receba os 3 lados pelo `input()` e:

1. Verifique se os lados formam um triângulo válido *(a soma de dois lados quaisquer deve ser maior que o terceiro)*
2. Se válido, classifique:
   - **Equilátero:** todos os lados iguais
   - **Isósceles:** dois lados iguais
   - **Escaleno:** todos diferentes

Exiba a classificação ou `"Não forma um triângulo"`.

````python
lado1 = float(input("Digite o primeiro lado do triângulo: "))
lado2 = float(input("Digite o segundo lado do triângulo: "))
lado3 = float(input("Digite o terceiro lado do triângulo: "))

if lado1 != 0 and lado2 != 0 and lado3 != 0:
  if lado1 == lado2 == lado3:
    print("Triângulo equilátero.")
  elif lado1 == lado2 or lado1 == lado3 or lado2 == lado3:
    print("Triângulo isósceles.")
  else:
    print("Triângulo escaleno.")
else:
  print("Não forma um triângulo.")
````

### 💡 Dica:

*Utilize somente se necessário — tente resolver primeiro sozinho!*

*Condição de triângulo válido: `a+b>c AND a+c>b AND b+c>a`. Depois compare os lados com `==` para classificar.*
